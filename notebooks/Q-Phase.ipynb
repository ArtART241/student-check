## Student Check: Question-Phase

Nach einer langwierigen Recherche habe ich mich dazu entschlossen, mich mit dem Thema von Studenten und Schülern zu beschäftigen. Konkreter gehe ich in diesem Projekt darauf ein, welche Attribute eines Schülers/Studenten auf seine Leistung in der Bildung Einfluss haben.

Ich stelle hiermit die Forschungsfrage auf: *Welche personenspezifische Merkmale haben den größtmöglichen Einfluss auf die Leistung von Schülern und Studenten, auf die dieselben eigenen Einfluss nehmen können?* Damit grenze ich Merkmale aus wie Alter, Geschlecht, ethnische Hintergründe und andere Voraussetzungen, auf die der Mensch selbst keinen Einfluss hat.

Ziel dieses Projekts ist es eine interaktive Streamlit Applikation zu bauen, wo nach Eingabe von eigenen Merkmalen ein Machine Learning Modell Empfehlungen aussprechen kann, welche Merkmale die Person hochziehen kann, um ihre Leistung auf ihrem Bildungsweg zu steigern. Diese Arbeit ist zwar zu diesem Zeitpunkt ergebnisoffen, aber man kann annehmen, dass beispielsweise eine Empfehlung sein könnte 12% mehr zu schlafen, um die notenbasierte Leistung um 30% zu verbessern.

In der nächsten Phase werden die gefundenen Datensätze genauer überprüft und es wird analysiert, mit welchen Machine Learning Modellen das Ziel erreicht werden kann, und was die Limitationen sind. Vorzugsweise werden Machine Learning Modelle betrachtet werden, die einen nachvollziehbaren "Gedankengang" haben (Explainable AI).
